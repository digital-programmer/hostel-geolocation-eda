```python
import numpy as np
import pandas as pd
import os
pd.set_option('display.max_rows', None)
```

```python
path = "food_coded.csv"
base_path = os.getcwd()
file_path = os.path.join(base_path,path)
file_path
```

```python
student_data = pd.read_csv(file_path)
student_data
```

```python
data = student_data.loc[:, ["cook","eating_out","employment","ethnic_food","exercise","fruit_day","income","on_off_campus","pay_meal_out","sports","veggies_day"]]
data
```

```python
nan_values = data.isna()
nan_columns = nan_values.any()
columns_with_nan = data.columns[nan_columns].tolist()
columns_with_nan 
```

```python
for col in columns_with_nan:
    mean_value = data[col].mean(skipna=True)
    data[col].fillna(float(round(mean_value)),inplace=True)
data
```
